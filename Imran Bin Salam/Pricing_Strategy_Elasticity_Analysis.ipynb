# Pricing Strategy & Elasticity Analysis

## 1. Dataset Creation

```

import pandas as pd
import numpy as np

np.random.seed(42)
price = np.linspace(10, 100, 50)
base_demand = 1000
demand = base_demand - 8 * price + np.random.normal(0, 30, len(price))
revenue = price * demand

data = pd.DataFrame({
    "Price": price,
    "Demand": demand,
    "Revenue": revenue
})

data.head()
```

## 2. Price vs Demand Visualization

```

import matplotlib.pyplot as plt

plt.figure()
plt.scatter(data["Price"], data["Demand"])
plt.xlabel("Price")
plt.ylabel("Demand")
plt.title("Price vs Demand")
plt.show()
```

## 3. Revenue Optimization Curve

```

plt.figure()
plt.plot(data["Price"], data["Revenue"])
plt.xlabel("Price")
plt.ylabel("Revenue")
plt.title("Price vs Revenue")
plt.show()
```

## 4. Elasticity Calculation (Log-Log Model)

```

data["log_price"] = np.log(data["Price"])
data["log_demand"] = np.log(data["Demand"])

elasticity = np.polyfit(data["log_price"], data["log_demand"], 1)[0]
elasticity
```


## Interpretation

If elasticity < -1 → Demand is Elastic  
If elasticity between -1 and 0 → Demand is Inelastic  

This helps determine optimal pricing strategy.
