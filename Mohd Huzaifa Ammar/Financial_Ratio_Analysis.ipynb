# Financial Performance & Ratio Analysis
This notebook performs profitability, liquidity, and efficiency ratio analysis.

## Step 1: Create Sample Financial Dataset

```

import pandas as pd

data = {
    "Year": [2019, 2020, 2021, 2022, 2023],
    "Revenue": [500000, 550000, 620000, 700000, 820000],
    "COGS": [300000, 320000, 360000, 390000, 450000],
    "Operating_Expenses": [80000, 85000, 90000, 100000, 115000],
    "Net_Income": [70000, 85000, 110000, 150000, 190000],
    "Current_Assets": [200000, 210000, 240000, 260000, 300000],
    "Current_Liabilities": [120000, 130000, 140000, 150000, 170000],
    "Inventory": [50000, 52000, 60000, 65000, 70000],
    "Total_Assets": [600000, 650000, 720000, 800000, 900000]
}

df = pd.DataFrame(data)
df
```

## Step 2: Calculate Financial Ratios

```

df["Gross_Profit"] = df["Revenue"] - df["COGS"]
df["Gross_Margin"] = df["Gross_Profit"] / df["Revenue"]
df["Net_Profit_Margin"] = df["Net_Income"] / df["Revenue"]
df["ROA"] = df["Net_Income"] / df["Total_Assets"]

df["Current_Ratio"] = df["Current_Assets"] / df["Current_Liabilities"]
df["Quick_Ratio"] = (df["Current_Assets"] - df["Inventory"]) / df["Current_Liabilities"]

df["Asset_Turnover"] = df["Revenue"] / df["Total_Assets"]
df["Inventory_Turnover"] = df["COGS"] / df["Inventory"]

df
```

## Step 3: Visualization Example

```

import matplotlib.pyplot as plt

plt.figure()
plt.plot(df["Year"], df["Gross_Margin"])
plt.title("Gross Margin Trend")
plt.xlabel("Year")
plt.ylabel("Gross Margin")
plt.show()
```
